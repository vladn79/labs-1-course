```python
df = pd.read_csv('weather.csv')
```

```python
df
```

```python
df.rename(columns={'хмарність': 'хмарність, %'}, inplace=True)
df.rename(columns={'опади': 'опади, м.м.'}, inplace=True)
df.rename(columns={'денна температура повітря': 'денна температура повітря, °C'}, inplace=True)
df.rename(columns={'нічна температура повітря': 'нічна температура повітря, °C'},inplace=True )
df.rename(columns={'сила вітру': 'сила вітру, м/с'}, inplace=True)
df['хмарність, %'] = df['хмарність, %'].replace('%', '', regex=True)
df['опади, м.м.'] = df['опади, м.м.'].replace('м.м.', '', regex=True)
df['опади, м.м.'] = df['опади, м.м.'].replace('-', '0', regex=True)
df['денна температура повітря, °C'] = df['денна температура повітря, °C'].replace('°C', '', regex=True)
df['нічна температура повітря, °C'] = df['нічна температура повітря, °C'].replace('°C', '', regex=True)
df['сила вітру, м/с'] = df['сила вітру, м/с'].replace('м/с', '', regex=True)
```

```python
df
```

```python
filtered_df = df[(df['період'] == '2022-09')]
filtered_df.loc[:, 'денна температура повітря, °C'] = filtered_df['денна температура повітря, °C'].astype(float)
filtered_df.loc[:, 'нічна температура повітря, °C'] = filtered_df['нічна температура повітря, °C'].astype(float)
filtered_df.loc[:, 'хмарність, %'] = filtered_df['хмарність, %'].astype(int)
filtered_df.loc[:, 'сила вітру, м/с'] = filtered_df['сила вітру, м/с'].astype(int)
filtered_df.loc[:, 'опади, м.м.'] = filtered_df['опади, м.м.'].astype(float)
filtered_df
```

```python
plt.figure(figsize=(15, 12))

plt.subplot(2, 2, 1)
plt.plot(filtered_df['день'], filtered_df['денна температура повітря, °C'], color = "Red")
plt.xlabel('День')
plt.ylabel('Денна температура, °C')
plt.title('Зміна денної температури в Період 2022-09(вересня 2022 року)')

plt.subplot(2, 2, 2)
plt.plot(filtered_df['день'], filtered_df['нічна температура повітря, °C'], color = "Green")
plt.xlabel('День')
plt.ylabel('Hічна температура, °C')
plt.title('Зміна нічної температури в Період 2022-09(вересня 2022 року)')

plt.subplot(2, 2, 3)
plt.plot(filtered_df['день'], filtered_df['хмарність, %'], color = "#073163")
plt.xlabel('День')
plt.ylabel('хмарність, %')
plt.title('Зміна хмарності в Період 2022-09(вересня 2022 року)')

plt.subplot(2, 2, 4)
plt.plot(filtered_df['день'], filtered_df['сила вітру, м/с'], color = "#dba21d")
plt.xlabel('День')
plt.ylabel('сила вітру, м/с')
plt.title('Зміна сили вітру в Період 2022-09(вересня 2022 року)')

plt.tight_layout()
plt.show()


```

```python
plt.figure(figsize=(8, 6))
diameters = filtered_df['опади, м.м.'].apply(lambda x: 70*x if x > 0 else 5)

plt.scatter(filtered_df['день'], filtered_df['денна температура повітря, °C'], s=diameters)
plt.xlabel('День')
plt.ylabel('Денна температура, °C')
plt.title('Залежність денної температури від кількості опадів')

plt.show()
```

```python
plt.figure(figsize=(8, 8))


df['опади, м.м.'] = df['опади, м.м.'].astype(float)


grouped_df = df[df['опади, м.м.'] > 0].groupby(df['період'])['період'].count()

days_with_precipitation = grouped_df.values

patches, texts, autotexts = plt.pie(days_with_precipitation, labels=sorted(set(df['період'])), autopct='%1.0f%%')
total_days = sum(days_with_precipitation)
for i, text in enumerate(autotexts):
    percentage = (days_with_precipitation[i] / total_days) * 100
    text.set_text(f" {percentage:.1f}% \n {days_with_precipitation[i]}")


plt.legend(patches, set(df['період']), loc='upper right', bbox_to_anchor=(1.3, 1))
plt.subplots_adjust(top=0.95)

plt.axis('equal')
plt.title('Кругова діаграма кількості днів з опадами в кожному місяці \n (в середині кожного сектора діаграми відображається загальна відсоткава кількість та фактична кількість опадів)')

plt.show()


```

```python

df['хмарність, %'] = df['хмарність, %'].astype(float)

def determine_day_type(cloudiness):
    if cloudiness > 70:
        return 'Хмарний'
    elif cloudiness < 35:
        return 'Сонячний'
    else:
        return 'Мінлива хмарність'

df['тип дня'] = df['хмарність, %'].apply(determine_day_type)

grouped_df = df.groupby(['період', 'тип дня'])['тип дня'].count().unstack().fillna(0)

ax = grouped_df.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#dba21d', 'Red', '#3498eb'])

ax.set_xlabel('Місяці або періоди')
ax.set_ylabel('')
ax.set_title('Кількість днів різних типів за місяцем(цифри в середині діаграми - відповідні кількості днів різних типів для кожного місяця)')
for i, month in enumerate(grouped_df.index):
    total = 0
    for j, value in enumerate(grouped_df.iloc[i]):
        if value > 0:
            ax.text(i, total + value / 2, f"{int(value)}", ha='center', va='center')
            total += value

ax.legend(title='Тип дня', loc='upper right', bbox_to_anchor=(1.3, 1))

plt.show()

```

```python
plt.figure(figsize=(8, 6))
df['нічна температура повітря, °C'] = df['нічна температура повітря, °C'].astype(float)

df['денна температура повітря, °C'] = df['денна температура повітря, °C'].astype(float)


night_temp_deviation =abs(df['денна температура повітря, °C']- df['нічна температура повітря, °C']) 


plt.hist(night_temp_deviation, bins=20, color='#dba21d', edgecolor='black')


plt.xlabel('Відхилення температури, °C')
plt.ylabel('Кількість днів')
plt.title('Гістограма відхилення нічної температури від денної')

plt.show()
```
