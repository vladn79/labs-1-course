```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Task 2

```python
graph = nx.read_edgelist("graph.txt", delimiter=" ", nodetype=int)
pos = nx.spring_layout(graph, k=1) 
nx.draw_networkx_edges(graph, pos, edge_color="black", width=1)
nx.draw_networkx_nodes(graph, pos, node_size=300, node_color="white", edgecolors="black", linewidths=1)
nx.draw_networkx_labels(graph, pos, font_size=10)
plt.axis('off')
plt.savefig("task2.jpg")
plt.show()
```

# Task 3

```python
pos = {
    1: (-5, 3),
    2: (-5, -2),
    3: (-4, 3),
    4: (-4, -2),
    5: (-3, 3),
    6: (-3, -2),
    7: (-2, -2),
    8: (6, -2),
    9: (2, 3),
    10: (2, -1),
    11: (2, 1),
    12: (2, 2)
}

nx.draw_networkx_edges(graph, pos, edge_color="black", width=1)
nx.draw_networkx_nodes(graph, pos, node_size=300, node_color="white", edgecolors="black", linewidths=1)
nx.draw_networkx_labels(graph, pos, font_size=10)


plt.axis('off')
plt.savefig("task3.jpg")
plt.show()
```

# Task 4

```python
components = nx.connected_components(graph)
for i in components:
    component_graph = graph.subgraph(i)
    num_nodes = component_graph.number_of_nodes()
    num_edges = component_graph.number_of_edges()
    degrees = dict(component_graph.degree())
    eccentricities = nx.eccentricity(component_graph)
    radius = nx.radius(component_graph)
    diameter = nx.diameter(component_graph)
    print(f"A connectivity component consisting of such nodes:{i}")
    print("Number of nodes:", num_nodes)
    print("Number of edges:", num_edges)
    print("Degrees of nodes:", degrees)
    print("Eccentricities of nodes:", eccentricities)
    print("Radius:", radius)
    print("Diametr:", diameter)
    print("******************************************************************")
```

# Task 5

```python
components = nx.connected_components(graph)
for i in components:
    component_graph = graph.subgraph(i)
    d = nx.diameter(component_graph)
    d_path = None
    for node in i:
        paths = nx.single_source_dijkstra_path(component_graph, node)
        for target, path in paths.items():
            if len(path) == d + 1:
                d_path = path
                break
        if d_path is not None:
            for j in range(len(d_path) - 1):
                a = d_path[j]
                b = d_path[j + 1]
                component_graph.edges[a, b]['color'] = 'red'
                component_graph.nodes[a]['color'] = '#f5e642'
                component_graph.nodes[b]['color'] = '#f5e642'
            break
    for node in component_graph.nodes:
        if node not in d_path:
            component_graph.nodes[node]['color'] = 'white'
    nx.draw_networkx_nodes(component_graph, pos, node_color=[component_graph.nodes[node].get("color", "white") for node in component_graph.nodes()], node_size=300, edgecolors="black", linewidths=1)
    for a, b, color in component_graph.edges.data('color'):
        nx.draw_networkx_edges(component_graph, pos, edgelist=[(a, b)], edge_color=color, width=1)
    nx.draw_networkx_labels(graph, pos, font_size=10)
plt.axis('off')
plt.savefig("task5.jpg")
plt.show()
```
